## Bodyfat data set

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

file_path = 'http://www.andrew.cmu.edu/user/dvaroday/14757/data/lecture/BodyFat.xls'
df = pd.read_excel(file_path)
display(df)
```

```
df = pd.read_excel(file_path, usecols=[4, 5, 6])
df
```

### Interactive 3D scatter plot

```
import plotly.express as px
fig = px.scatter_3d(df, x='WEIGHT', y='HEIGHT', z='ADIPOSITY')
fig.show()
```

### Scatter plot matrix

```
import seaborn as sns
sns.pairplot(df)
```

### Principal components analysis on standardized data

```
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
df_no_outlier = df[df.HEIGHT > 40] # remove height outlier
data_standardized = StandardScaler().fit_transform(df_no_outlier)
pca = PCA(n_components=3)
principal_components = pca.fit_transform(data_standardized)
df_principal = pd.DataFrame(data = principal_components, columns = ['pc1', 'pc2', 'pc3'])

fig = px.scatter_3d(df_principal, x='pc1', y='pc2', z='pc3')
fig.show()
```

### Projection of data to 2 dimensions

```
df_projected = df_principal
df_projected.pc3 = 0 # zero out principal component 3

fig = px.scatter_3d(df_projected, x='pc1', y='pc2', z='pc3')
fig.show()
```
